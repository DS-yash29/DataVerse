```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=7)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
class LogisticRegression():
    
    def __init__(self):
        self.X = None
        self.y = None
        self.w = None
        
    def step(z):
        return 1 if z>0 else 0
        
    def percept(self,X,y):
        self.X = X
        self.y = y
    
        X = np.insert(X,0,1,axis=1)
        self.w = np.ones(X.shape[1])
        lr = 0.1

        for i in range(1000):
            j = np.random.randint(100)
            y_hat = step(np.dot(X[j],self.w))
            self.w = self.w + lr*(y[j]-y_hat)*X[j] 
            
        
        return None
    
    def predict(self):
        X_ = np.insert(self.X,0,1,axis=1)
        out = np.ones(X_.shape[0])
        for i in range(X_.shape[0]):
            out[i] = step((np.dot(X_[i],self.w)))
        return out
    
    def computeAccuracy(self):
        out = self.predict()
        
        diff = 0
        for i in range(len(self.y)):
            diff += abs(y[i]-out[i])
        
        return round(1- (diff/len(self.y) ), 2)
```

```python
lr1 = LogisticRegression()
```

```python
lr1.percept(X=X,y=y)
```

```python
lr1.predict()
```

```python
lr1.computeAccuracy()
```
