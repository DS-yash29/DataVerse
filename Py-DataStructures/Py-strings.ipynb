**Strings Creation**

Through:
1. 'Single quotes'
2. "Double Quotes"
3.  " " "Multiple Quotes" " " -> For Strings consisting of Multiple lines.


```python
s1 = 'Hi My Name is Yash '
s2 = "I Am an Aspiring Data Scientist"
s3 = '''\nI Like Exploring New Stuff,
I Like Travelling,
Listening to Music &
Spending time in Nature'''
s4 = """\nHello in French is
Bonjour"""
print(s1,s2,s3,s4,sep='\n')
```

```python
# Indexing

print(s1[0]) # Indexing postion starts form 0 -> length -1
print(s1[-1]) # -> Negatice Indeixng
# print(s1[35]) -> Index out of range
print(s1[-3])
#print(s1[-29]) ->Index out of Range
```

```python
# Slicing
print(s2[0:9:]) # STring Getting printed till 8th Index
print(s2[8::-1]) # Reverse Printing / Negative Slicing
print(s2[-6:-11:-1])
```

# String Functions

**`string_variable. -> shows all the functions related to strings`**


### Common Functions :
- len / min / max / sorted

### Cases in Strings
- Uppercase/ Capitalize / LowerCase / Title / Swapcase

### Position related Functions
- count / find / index

### Content Related Functions

- endswith / startswith / isalpha / isNumberic / isalphanumeric / isdigit / isindentifier

### Splitting & Joining Based
- split / join / replace / strip

```python
# Common Functions
print(len(s1))
print(min(s1))
print(max(s1))
print(sorted(s1,reverse=True))
```

```python
# Cases Functions
s = 'hi my name is yash'
print(s1.capitalize())
print(s1.upper())
print(s1.lower())
print(s1.title())
print(s1.swapcase())
print(s1.casefold())
```

```python
# Position Related Functions
print(s.count('a'))
print(s.find('m'))
print(s.index('h'))

# Inserting Elements which doesnt exists to see the difference between find & index Fnctions
print(s.find('u')) # Find returns -1 when the character we are searching doest not exixsts
#print(s.index('u')) # Index generates error for the same
print(s.find('my'))
print(s.count('yash'))
```

```python
# Content related functions
print(s.startswith('m'))
print(s.endswith('h'))
print(s.isalnum())
print(s.isalpha())
print(s.isidentifier())
print(s.isupper())
print(s.isprintable())
print(s.isascii())
print(s.islower())
```

```python
# Splitting & Joining
a = s.split()
print(a)

s = " ".join(a)
print(s)
```

```python
# Replace
s.replace('yash','yash 2.0')
```

```python
# STrip
x = " yash       ".strip()
print(x)
```
