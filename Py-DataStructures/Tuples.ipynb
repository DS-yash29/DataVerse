*Tuples*

```python
# Creation
a = () # Empty Tuple
print(a,type(a))

b = (1,2,3,4,5) # 1-d Homogeneous Tuples
print(b,type(b),'\n')

c = (1,"Hello",23.7,True,(3,4,5,6),["Yes","No",2435])
for i in c:
  print(i,type(i),end='\n')

d = (1,) # A Tuple containg single set must be created by adding , folllowed by the first element otherwise the type of the elemeent will be the type of the variable
e = (1) # e will be of type int and not tuple
print(d,e,type(d),type(e),sep='  -  ')

```

```python
# Indexing & Slicing
b = (1,2,3,4,5)
c = (1,"Hello",23.7,True,(3,4,5,6),["Yes","No",2435])

print(b[:]) # By default entire tuple would be printed
print(b[0:-1]) # Tuple sliced from 0th index to the last element(excluded)
print(c[-1:-4:-1]) # Negative Slicing Backwards
print(c[-5:-1]) #
```

```python
# Manipulation of Tuples
b = (1,2,3,4,5)

#del b[0] -> Not Allowed : TypeError: 'tuple' object doesn't support item deletion
#b[0] = "Hey" -> Not Allowed : TypeError: 'tuple' object does not support item assignment

# del b -> deletes the tuple
```

*Tuples in Python are Immutable*

## Operations in Tuples:

- Arithmetic
- Membership
- Loops

```python
b = (1,2,3,4,5)
d = ("Yash",29,27.49,True)
# Multiplication
(print(b*2))

# Membership
print("Yash" in d)

# Loops
for i in d:
  print(i,end=' ')
```

## Functions in Tuples:

- Basic Functions -> len/min/max/common

- Position related functions -> count/index/find

```python
d = (5,-6,1,445,34,-9348,34)

print(len(d))
print(min(d))
print(max(d))
print(sorted(d,reverse=False))
print(sorted(d,reverse=True))
```

```python
print(d.index(34))
print(d.count(8))
```

Difference between Lists and Tuples

    Syntax -> Similar
    Mutability   -> Lists Are Mutable | Tuples are not
    Speed   -> Tuples are generally faster than Lists
    Memory   -> Lists Consumes more memory than Tuples
    Built in functionality   -> Lists have considerably more functions
    Error prone
    Usability


```python
import sys

l = list(range(100000000))
t = tuple(range(100000000))

print(sys.getsizeof(l),sys.getsizeof(t),sep='\n')
```

```python
import time

start = time.time()
for i in l:
  i*2
print("Time taken to traverse: ",(time.time()-start)*10)

start = time.time()

for i in t:
  i*2
print("Time taken to traverse: ",(time.time()-start)*10)
```
